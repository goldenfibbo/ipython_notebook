```python
df=pd.read_csv('avocado.csv')
df['Date']=pd.to_datetime(df['Date'])
albany_df=df.copy()[df['region']=='Albany']
albany_df.set_index('Date',inplace=True)
albany_df['AveragePrice'].plot()

dff=pd.read_csv('avocado.csv')
#dff=dff.copy()[dff['type'] == 'organic']
albany_dff=dff.copy()[dff['region']=='Albany']
albany_dff.sort_values(by='Date',ascending=True,inplace=True)
albany_dff[['AveragePrice','Date']]
albany_dff.plot('Date','AveragePrice')
```
