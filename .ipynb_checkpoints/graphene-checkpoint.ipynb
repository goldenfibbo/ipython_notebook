```python
import numpy as np
import re
import matplotlib.pyplot as plt
import pandas as pd
import os
pd.set_option('display.max_rows', None)
```

# plot the scattering phase space

```python
df = pd.read_csv("graphene.sps",sep="\s+")
```

```python
df = df.set_index("k")
```

```python
a = np.array(df["P+"])
b = np.array(df["P-"])
```

```python
df["P3"] = (2*a+b)*301**2
```

```python
df1 = df[df["mode"]==1]
df2 = df[df["mode"]==2]
df3 = df[df["mode"]==3]
```

```python
df3[abs(df3["omega"]*0.03-27.01386)<=0.00001]
```

```python
omega1 = np.array(df1["omega"])
omega2 = np.array(df2["omega"])
omega3 = np.array(df3["omega"])
p31 = np.array(df1["P3"])
p32 = np.array(df2["P3"])
p33 = np.array(df3["P3"])
omega1 = omega1*3*0.01
omega2 = omega2*3*0.01
omega3 = omega3*3*0.01
```

```python
fig = plt.figure(figsize=(20,20))
#ax = plt.gca()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax1.scatter(omega1[1:],p31[1:],s=20,c="magenta")
ax2.scatter(omega2[1:],p32[1:],s=20,c="cyan")
ax3.scatter(omega3[1:],p33[1:],s=20,c="blue")
ax1.set_xlabel("Freq/THz")
ax1.set_ylabel(r"$P_3$")
ax2.set_xlabel("Freq/THz")
ax2.set_ylabel(r"$P_3$")
ax3.set_xlabel("Freq/THz")
ax3.set_ylabel(r"$P_3$")
ax1.tick_params(labelsize=17)
ax2.tick_params(labelsize=17)
ax3.tick_params(labelsize=17)
ax1.set_yscale("log")
ax2.set_yscale("log")
ax3.set_yscale("log")
plt.savefig("fig.jpg")
```
