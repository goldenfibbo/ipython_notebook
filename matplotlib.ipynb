ordinary plotting

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
x2 = [3,2,1]
y2 = [1,2,3]
plt.plot(x,y,'-.',label='proportional')
plt.plot(x2,y2,'--',label='anti-proportional')
plt.title(r'\LaTeX,$\displaystyle\oint x$')
#plt.title(r'\TeX\ is Number $\displaystyle\sum_{n=1}^\infty'
#             r'\frac{-e^{i\pi}}{2^n}$!', fontsize=16, color='r')
plt.xlabel(r'$\displaystyle\frac{x}{a}$')
plt.ylabel('y')
plt.legend()
#plt.savefig(fname='testfig',format='pdf')
plt.show()
```

stack plot
---

```python
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating =   [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]

plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.show()
```

```python
plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r', label='Working', linewidth=5)
plt.plot([],[],color='k', label='Playing', linewidth=5)
#no actual plotting, just to add legends
plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

pie charts
---

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow= True,
        explode=(0,0.1,0,0),#highlight some part out of the whole circle
        autopct='%1.1f%%')

plt.title('Interesting Graph\nCheck it out')
plt.show()
```

```python
import csv
import pandas as pd
```

```python
#alternative way by pandas
df = pd.read_csv('example.txt',index_col=0,header=None)
df.rename(columns={1:'values',0:'index'},inplace=True)
#df.columns = ['value']
df.reset_index().columns
df.plot()
plt.show()
```

```python
x=[]
y=[]

with open('example.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

plt.plot(x,y, label='Loaded from file!')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

```python
x, y = np.loadtxt('example.txt', delimiter=',', unpack=True)
plt.plot(x,y, label='Loaded from file!')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

```python
import urllib
import matplotlib.dates as mdates
```

```python
def graph_data(stock):

    # Unfortunately, Yahoo's API is no longer available
    # feel free to adapt the code to another source, or use this drop-in replacement.
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'

    source_code = urllib.request.urlopen(stock_price_url).read().decode()

    stock_data = []
    split_source = source_code.split('\n')

    for line in split_source[1:]:
        split_line = line.split(',')
        if len(split_line) == 7:
            if 'values' not in line:
                stock_data.append(line)
```

```python
graph_data('AAPL')
```
